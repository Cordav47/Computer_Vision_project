```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from Additional_functions import Imp
```

```python
eucerin = Imp("Photos/Extras/Eucerin.jpg")[3]

D3 = Imp("Photos/Toothpaste/D3.jpg")[3]

he = np.shape(eucerin)[0]
we = np.shape(eucerin)[1]
eucerin = cv2.rotate(eucerin, cv2.ROTATE_180)
bh = 2100
bw = 1600
D3[bh: bh + he, bw: bw+we] = eucerin

plt.imshow(D3)
plt.show()
D3 = cv2.cvtColor(D3, cv2.COLOR_BGR2RGB)
cv2.imwrite("D10.jpg", D3)
```

Now create an analog image for the tissue with the written flipped. I will use the dst found from the Tissue_setup.ipynb file

```python
#955 to 1525 and 2530 to 2700

TIS = Imp("Photos/Tissue/S1.jpg")[3]

written = TIS[2530:2700, 955:1525]
upside_written = cv2.rotate(written, cv2.ROTATE_180)

plt.imshow(upside_written)
plt.show()
```

Then, we create two images, one where the written is flipped and another one where it is misplaced. Plot to see if they come out correctly and then save them

```python
upside = np.copy(TIS)
misplaced = np.copy(TIS)

upside[2530:2700,955:1525] = upside_written
plt.imshow(upside)
plt.show()
misplaced[2000:2000+2700-2530, 1200:1200+1525-955] = written
orange_color = np.mean(TIS[2000:2100, 1500:1600], axis = 1)
orange_color = np.int32(orange_color)
misplaced[2530:2700,955:1525] = orange_color[0]
plt.imshow(misplaced)
plt.show()

#And finally save them
upside = cv2.cvtColor(upside, cv2.COLOR_BGR2RGB)
cv2.imwrite("S8.jpg", upside)
misplaced = cv2.cvtColor(misplaced, cv2.COLOR_BGR2RGB)
cv2.imwrite("S9.jpg", misplaced)
```
